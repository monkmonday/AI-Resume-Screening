```python
import re
import nltk
from nltk.corpus import stopwords

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)
```

```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text
```

```python
resume = extract_text_from_pdf(r"C:\Users\MOHIT JHA\Downloads\Mohit_Jha_Resume_ML.pdf")

job_desc = """Looking for a data scientist with Python, machine learning,
data analysis, SQL, and deep learning experience"""
```

```python
resume_clean = preprocess(resume)
job_clean = preprocess(job_desc)
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,2))
vectors = vectorizer.fit_transform([resume_clean, job_clean])
```

```python
similarity = cosine_similarity(vectors[0], vectors[1])
score = similarity[0][0]

print("Match Score:", round(score * 100, 2), "%")
```

```python
skills = {
    "python": ["python"],
    "machine learning": ["machine learning", "ml"],
    "data analysis": ["data analysis", "data analytics"],
    "deep learning": ["deep learning"],
    "nlp": ["nlp", "natural language processing"],
    "tensorflow": ["tensorflow"],
    "sql": ["sql"],
    "pandas": ["pandas"]
}

found_skills = []
missing_skills = []

for skill, keywords in skills.items():
    if any(keyword in resume_clean for keyword in keywords):
        found_skills.append(skill)
    else:
        missing_skills.append(skill)

print("Skills Found:", found_skills)
print("Missing Skills:", missing_skills)
```

```python
if score > 0.75:
    print("Excellent Match")
elif score > 0.5:
    print("Good Match")
elif score > 0.3:
    print("Average Match")
else:
    print("Low Match")
```

```python
if missing_skills:
    print("\nRecommended Skills to Learn:")
    for skill in missing_skills:
        print("-", skill)
```
